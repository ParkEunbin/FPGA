# FPGA

- 검증용: ASIC 제품을 만들기 위해 검증하기 위해서 설계한 IP를 올려보는 것
- 제품용: FPGA자체로 제품을 만듬(수량이 적은 경우)


### FPGA 설계 FLOW
image.png

1. System Design Entry : IP를 설계하고 선정하는 단계. 직접 verilog 및, HLS 등으로 IP를 만들 수 있음.HW적인 IP를 준비

2. Implementation: FPGA fitting / 설계한 IP를 FPGA에 올리고 동작시키기 위해서 진행하는 작업/ testbench를 만들어 function적으로 문제가 없는지 확인> FPGA에 올리기 위해 필요한 constraints들을 기술 (clock 속도, 핀 배치 등등을 .xdc 파일에 기술) > 합성> implementation(PnR) > Timing closure가 되엇는가, 누락된 routing은 없는가, resource가 부족하진 않은가 등 design을 분석> FPGA에 올릴 수 있는 Bitstream 파일을 생성할 수 있음 

3. HW Bring-UP and Validation & Software Development : FPGA에 Bitstream을 오려서 동작을 검증 및 debugging을 함 / 동작시키기 위해서는 대부분 SW가 필요함.
